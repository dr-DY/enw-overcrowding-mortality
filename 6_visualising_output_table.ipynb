```python
import pandas as pd

# Save final_df to formatted Excel
output_path = "Output/Predictions/Formatted_Prison_Deaths_Projection.xlsx"

with pd.ExcelWriter(output_path, engine='xlsxwriter') as writer:
    final_df.to_excel(writer, sheet_name='Deaths')

    workbook = writer.book
    worksheet = writer.sheets['Deaths']

    # Formats
    header_fmt = workbook.add_format({'bold': True, 'align': 'center', 'valign': 'vcenter', 'border': 1})
    index_fmt = workbook.add_format({'align': 'left', 'valign': 'vcenter', 'border': 1})
    cell_fmt = workbook.add_format({'align': 'center', 'valign': 'vcenter', 'border': 1})

    # Adjust column widths
    worksheet.set_column(0, 0, 25)  # Death Category
    worksheet.set_column(1, 1, 30)  # Prison Type
    worksheet.set_column(2, final_df.shape[1] + 1, 20)  # Year columns

    # Manually rewrite headers with formatting
    worksheet.write(0, 0, 'Death Category', header_fmt)
    worksheet.write(0, 1, 'Prison Type', header_fmt)

    for col_idx, col_name in enumerate(final_df.columns, start=2):
        worksheet.write(0, col_idx, col_name, header_fmt)

    # Write MultiIndex rows with formatting
    for row_idx, (death_cat, prison_type) in enumerate(final_df.index, start=1):
        worksheet.write(row_idx, 0, death_cat, index_fmt)
        worksheet.write(row_idx, 1, prison_type, index_fmt)

        for col_idx, col_name in enumerate(final_df.columns, start=2):
            worksheet.write(row_idx, col_idx, final_df.loc[(death_cat, prison_type), col_name], cell_fmt)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Updated 2029 data using Other_inclHomicide
data = [
    ['Category C', 37741.45751, '91 (64-124)', '17 (13-22)', '67 (46-92)', '9 (6-12)'],
    ['Category B', 30496.71536, '138 (96-189)', '53 (36-72)', '71 (49-97)', '14 (10-19)'],
    ['Category B + YOI', 8752.029699, '44 (12-81)', '19 (6-36)', '23 (6-42)', '5 (1-10)'],
    ['Category A', 5695.617787, '33 (13-60)', '8 (3-14)', '23 (9-41)', '3 (1-5)'],
    ['Category D', 5066.903398, '7 (3-11)', '1 (0-1)', '7 (3-11)', '1 (0-1)'],
    ['Category C + YOI', 4309.192156, '10 (3-20)', '2 (1-4)', '7 (2-15)', '1 (0-2)'],
    ['Young Offender Institution', 3472.455815, '11 (5-19)', '3 (1-5)', '8 (3-13)', '1 (1-2)'],
    ['Female (Closed)', 3195.867969, '11 (5-18)', '3 (1-4)', '8 (4-14)', '1 (0-2)'],
    ['Mixed [B + Female (Closed)]', 1437.559519, '5 (5-19)', '1 (1-4)', '3 (3-13)', '1 (1-2)'],
    ['Female (Closed) + YOI', 409.0711, '1 (1-4)', '0 (0-1)', '1 (1-3)', '0 (0-1)'],
    ['Female (Open)', 223.1296909, '1 (1-4)', '0 (0-1)', '1 (0-2)', '0 (0-0)'],
]
columns = ['PrisonType', 'Population', 'Deaths', 'SelfInflicted', 'Natural', 'Other']
df = pd.DataFrame(data, columns=columns)

# Extract numeric values from strings
for col in ['SelfInflicted', 'Natural', 'Other', 'Deaths']:
    df[col] = df[col].astype(str).str.extract(r'([\d.]+)').astype(float)

# Calculate per 10,000 inmates
for col in ['SelfInflicted', 'Natural', 'Other', 'Deaths']:
    df[col] = (df[col] / df['Population']) * 10000

# Round to nearest whole number
df = df.round({'SelfInflicted': 0, 'Natural': 0, 'Other': 0, 'Deaths': 0})

# Melt into long format
heatmap_long = df.melt(
    id_vars=['PrisonType'],
    value_vars=['SelfInflicted', 'Natural', 'Other', 'Deaths'],
    var_name='Death Type',
    value_name='Deaths per 10,000 Inmates'
)

# Rename columns
heatmap_long['Death Type'] = heatmap_long['Death Type'].replace({
    'SelfInflicted': 'Self-inflicted',
    'Natural': 'Natural causes',
    'Other': 'Other causes (incl. homicide)',
    'Deaths': 'All causes'
})

# Normalize and scale
heatmap_long["Normalized"] = heatmap_long["Deaths per 10,000 Inmates"] / heatmap_long["Deaths per 10,000 Inmates"].max()
heatmap_long["Size"] = np.sqrt(heatmap_long["Normalized"]) * 2500

# Labels
x_labels = ['Self-inflicted', 'Natural causes', 'Other causes (incl. homicide)', 'All causes']
y_labels = [
    'Category A', 'Category B', 'Category B + YOI', 'Category C', 'Category C + YOI',
    'Category D', 'Young Offender Institution', 'Female (Closed)',
    'Mixed [B + Female (Closed)]', 'Female (Closed) + YOI', 'Female (Open)'
]
heatmap_long["x"] = heatmap_long["Death Type"].apply(lambda x: x_labels.index(x))
heatmap_long["y"] = heatmap_long["PrisonType"].apply(lambda x: y_labels.index(x))

# Color map
custom_cmap = sns.color_palette("YlOrBr", as_cmap=True)

# Plot
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.1)
fig, ax = plt.subplots(figsize=(10, 7), facecolor='white')
ax.set_facecolor('white')

# Grid and frame
for spine in ax.spines.values():
    spine.set_visible(False)
ax.grid(axis='y', linestyle='--', alpha=0.3, color='gray')
ax.grid(axis='x', visible=False)

# Circles
scatter = ax.scatter(
    x=heatmap_long["x"],
    y=heatmap_long["y"],
    s=heatmap_long["Size"],
    c=heatmap_long["Normalized"],
    cmap=custom_cmap,
    edgecolors='white',
    linewidths=1.5,
    alpha=0.85,
    zorder=3
)

# Add numbers
for _, row in heatmap_long.iterrows():
    color = 'white' if row["Normalized"] > 0.5 else 'black'
    ax.text(row["x"], row["y"], f'{int(row["Deaths per 10,000 Inmates"])}',
            ha='center', va='center', fontsize=10, color=color)

# Axes
ax.set_xticks(range(len(x_labels)))
ax.set_xticklabels(x_labels, fontsize=11)
ax.set_yticks(range(len(y_labels)))
ax.set_yticklabels(y_labels, fontsize=11)
ax.tick_params(axis='both', which='both', length=0)
ax.set_xlim(-0.5, len(x_labels) - 0.5)
ax.set_ylim(-1, len(y_labels))  # Extra padding
ax.invert_yaxis()

# Add horizontal separators
for i in range(1, len(y_labels)):
    ax.axhline(y=i-0.5, color='lightgray', linestyle='-', alpha=0.3, zorder=1)

# Title and labels
plt.suptitle("Projected Yearly Death Rates per 10,000 Inmates for 2029", fontsize=18, weight='bold', y=0.98, color='#333333')
ax.set_xlabel("Cause of Death", fontsize=13, fontweight='bold', color='#333333', labelpad=12)
ax.set_ylabel("Prison Type", fontsize=13, fontweight='bold', color='#333333', labelpad=12)

plt.tight_layout()
plt.subplots_adjust(top=0.88, bottom=0.12)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set style and palette
sns.set_style('whitegrid')
plt.rcParams.update({'font.size': 12})
colors = sns.color_palette("colorblind")

# Load the data
file_path = 'Data/deaths_in_custody_by_prison.xlsx'
df = pd.read_excel(file_path)
df['date'] = pd.to_datetime(df['Date'])

# Group deaths by calendar year
df['year'] = df['Date'].dt.year
yearly_deaths = df.groupby('year')['incidents'].sum().sort_index()

# Projected values for 2025–2029
proj_years = [2025, 2026, 2027, 2028, 2029]
proj_means = [319, 341, 349, 357, 359]
proj_lows = [285, 300, 308, 314, 315]
proj_highs = [353, 381, 390, 402, 407]

# Combine observed and projected data
all_years = list(yearly_deaths.index) + proj_years
all_values = list(yearly_deaths.values) + proj_means

# Plot
plt.figure(figsize=(12, 6))

# Full connected line: observed + projected
plt.plot(all_years, all_values, marker='o', linestyle='-', color=colors[0], label='Observed')

# Dashed overlay for projected segment only
plt.plot(proj_years, proj_means, linestyle='-', color=colors[1], marker='o', label='Projected')

# Fill 95% CI for projections
plt.fill_between(proj_years, proj_lows, proj_highs, color=colors[1], alpha=0.3, label='95% CI (Projected)')

# Axes formatting
plt.title('Total Prison Deaths by Year with Projections (2025–2029)', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Deaths', fontsize=14)
plt.grid(True, alpha=0.3)

# Set x-axis to extend slightly before 2004 to prevent marker clipping
plt.xlim(2003.5, 2030)
xticks = list(range(2004, 2030))
plt.xticks(xticks, rotation=45)

# Set y-axis to start at 0
plt.ylim(bottom=0)

plt.legend()
plt.tight_layout()

# Save and show
plt.savefig('Output/Predictions/prison_deaths_projected_2025_2029.png', dpi=300, bbox_inches='tight')
plt.show()
```
